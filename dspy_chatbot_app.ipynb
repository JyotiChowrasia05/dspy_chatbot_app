```python
# signature include the task, input and output generated by LLM
import dspy
class ShakespeareanTransalation(dspy.Signature):
  """ Translate simple english to Shakespearsean English"""
  simple_english = dspy.InputField()
  shakkespearean_english = dspy.OutputField()
```

```python
from google.colab import userdata
OPEN_AI_KEY = userdata.get('gemini_api_key')
turbo = dspy.OpenAI(model = 'gpt-3.5-turbo', max_tokens = 1000, api_key= OPEN_AI_KEY)
```

```python
dspy.settings.configure(lm = turbo)
from dspy.signatures.signature import signature_to_template
shakespeare_transalation_as_template = signature_to_template(ShakespeareanTransalation)
```

```python
print(str(shakespeare_transalation_as_template))
```

```python
class CoT(dspy.Module):
  def __init__(self):
    super().__init__()
    self.prog = dspy.ChainOfThought(ShakespeareanTransalation)

  def forward(self, simple_english):
    return self.prog(simple_english = simple_english)

c = CoT
```
